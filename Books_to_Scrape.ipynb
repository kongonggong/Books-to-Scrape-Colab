random virtual IP Adress

```python
def randip():
  vip = socket.inet_ntoa(struct.pack('>I', random.randint(1, 0xffffffff)))
```

#Book Data Scarping

**Task : review stars⭐️, Book's name, Price, Status(in stock or out of stock) , img url**

```python
# Function to get articles from each page
def get_articles(page_num):
    base_url = base_url = 'https://books.toscrape.com/catalogue/page-{}.html'
    agent = {"User-Agent": 'Chrome/{}'.format(randip())}
    url = base_url.format(page_num)
    response = requests.get(url, headers=agent)
    response.raise_for_status()
    soup = BS(response.content, 'html.parser')
    return soup.find_all('article', {"class": "product_pod"})
```

```python
b_list=[]
page_num = 1

with tqdm(total=50) as pbar:##total page
  while True:
      try:
          articles = get_articles(page_num)
          if not articles:
              break
          b_list.extend(articles)
          page_num += 1
          pbar.update(1)#update bar
      except requests.exceptions.RequestException as e:
          print(f"Error fetching page {page_num}: {e}")
          break
```

```python
len(b_list)
```

```python
for i in b_list:#print first book to read a data
  print(i)
  break
```

```python
tt, st, p, sta, bu = ([] for i in range(5))
for article in tqdm(b_list):
    try:
        title = article.h3.a["title"]
        book_url = 'https://books.toscrape.com/catalogue/' + article.h3.a["href"]
        stars = article.p['class'][1]
        price = article.find('p', {"class": "price_color"}).text
        status = article.find('p', {"class": "instock availability"}).text.strip()

        #if you want to see a data while the code is running
        # print(title)
        # print(stars)
        # print(price)
        # print(status)
        # print(book_url)
        # print('========')

        tt.append(title)
        st.append(stars)
        p.append(price)
        sta.append(status)
        bu.append(book_url)
    except AttributeError as e:
        print(f"Error parsing article: {e}")
```

```python
dictt = {
    "Book's Title": tt,
    'Stars': st,
    'Price': p,
    'Status': sta,
    "Url": bu
}
```

```python
df = pd.DataFrame.from_dict(dictt)
```

```python
df.to_csv('result.csv', index=False)
```
